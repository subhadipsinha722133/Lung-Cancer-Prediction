```python
from keras import Sequential
from keras.layers import Dense 
```

```python
x_train.shape
```

```python
from keras.utils import to_categorical
y=to_categorical(y,3)
```

```python
from sklearn.model_selection import train_test_split 

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model=Sequential()

model.add(Dense(90,input_dim=23,activation="tanh"))
model.add(Dense(500,activation="sigmoid"))
model.add(Dense(500,activation="sigmoid"))
model.add(Dense(500,activation="relu"))
model.add(Dense(300,activation="sigmoid"))
model.add(Dense(2000,activation="relu"))
model.add(Dense(700,activation="tanh"))
model.add(Dense(800,activation="softmax"))
model.add(Dense(200,activation="sigmoid"))
model.add(Dense(4000,activation="sigmoid"))
model.add(Dense(200,activation="relu"))
model.add(Dense(7000,activation="tanh"))
model.add(Dense(8000,activation="softmax"))
model.add(Dense(200,activation="sigmoid"))
model.add(Dense(400,activation="sigmoid"))

model.add(Dense(3,activation="softmax"))

```

```python
model.summary()
```

```python

model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=60,validation_data=(x_test,y_test),batch_size=400)
```
